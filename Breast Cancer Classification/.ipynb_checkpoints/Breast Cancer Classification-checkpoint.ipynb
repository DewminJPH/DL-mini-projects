```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# Add target column in the data
data_frame['label']= breast_cancer_dataset.target
```

```python
data_frame.groupby('label').mean()
```

```python
# Separating the features and target
x = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

```python
  x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

# Building the Neural Network

```python
%pip install tensorflow
```

```python
# importing tenosorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
]) # here I have created a 1 input layer, 1 hidden layer and 1 output layer.
```

```python
# compiling the Neural Network
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# training the Neural Network
history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=10)
```

```python
# Visualizing accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='upper right')
```

```python
# Accuracy of the model on test data
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```

Building the predictive system

```python
input_data = (13.03,18.42,82.61,523.8,0.08983,0.03766,0.02562,0.02923,0.1467,0.05863,0.1839,2.342,1.17,14.16,0.004352,0.004899,0.01343,0.01164,0.02671,0.001777,13.3,22.81,84.46,545.9,0.09701,0.04619,0.04833,0.05013,0.1987,0.06169)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standarizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if (prediction_label[0]==0):
    print("The tumor is Malignant")
else:
    print('The tumor is Benign')
```
