```python
import numpy as np
import tensorflow as tf
```

```python
with open("MLprojects.txt", "r", encoding='utf-8') as file:
    mlprojects_details = file.read()
```

```python
mlprojects_details
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
```

```python
tokenizer.fit_on_texts([mlprojects_details])
```

```python
tokenizer.get_config()
```

```python
tokenizer.word_index
```

```python
input_sequences = []
for sentence in mlprojects_details.split('\n'):
    token_list = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(token_list)):
        sequence = token_list[:i+1]
        input_sequences.append(sequence)
```

```python
input_sequences[:5]
```

```python
# to find the what is the maximum length of sequence here
max_length = max([len(input_sequence) for input_sequence in input_sequences])
max_length
```

```python
# Now add the padding to other sequences with match to the maximum length
input_sequences = np.array(tf.keras.preprocessing.sequence.pad_sequences(input_sequences, maxlen=max_length, padding='pre'))
```

```python
input_sequences[0]
```

```python
x = input_sequences[:, :-1]
y = input_sequences[:, -1]
```

```python
x[0]
```

```python
y
```

```python
num_classes = len(tokenizer.word_index) + 1
num_classes
```

```python
# one hot encoding for all y values
y = np.array(tf.keras.utils.to_categorical(y, num_classes = num_classes))
```

```python
y[0]
```

# model initialization

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(num_classes, 80, input_length=max_length-1))
model.add(tf.keras.layers.LSTM(100))
model.add(tf.keras.layers.Dense(num_classes, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer = 'adam',
             metrics=['accuracy'])
```

```python
history = model.fit(x,y, epochs = 100)
```

```python
input_text = "Machine learing"
for i in range(6):
    token_list = tokenizer.texts_to_sequences([input_text])[0]
    token_list = tf.keras.preprocessing.sequence.pad_sequences([token_list], maxlen=max_length-1, padding='pre')
    prediction = np.argmax(model.predict(token_list), axis=1)
    output_word = ""
    for word, index in tokenizer.word_index.items():
        if index == prediction:
            output_word = word
            break
    input_text += " " + output_word

print(input_text)
```
