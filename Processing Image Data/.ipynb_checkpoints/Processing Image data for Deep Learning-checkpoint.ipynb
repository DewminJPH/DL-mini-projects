```python
import requests
from IPython.display import Image, display

url = 'https://media.vanityfair.com/photos/6224ec054471a41eee9742bc/master/w_2560%2Cc_limit/1238860697'
response = requests.get(url)

# 1. Save it with the correct .jpg extension
filename = 'Robert_pattinson.jpg'
with open(filename, 'wb') as f:
    f.write(response.content)

# 2. Display it in the notebook
display(Image(filename=filename)) 
```

# Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('Robert_pattinson.jpg')
img_resize = img.resize((1280,836))
```

```python
img_resize.save("Robert_pattinson_resized.jpg")
```

```python
# displaying the image from numpy array

img_res = mpimg.imread("Robert_pattinson_resized.jpg")
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

 # Converting RGB image into Grayscale image using OpenCV

```python
# importing OpenCV library
import cv2 
```

```python
img = cv2.imread('Robert_pattinson.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image. But this will not be allowed in Google Colab.

```python
# displaying the grayscale image
plt.imshow(grayscale_image, cmap='gray')
plt.show()
```

```python
# saving the grayscale image
cv2.imwrite('dog_grayscale_image.jpg', grayscale_image)
```
